```python
import pandas as pd
import numpy as np
import sys
import os 
```

```python
from sklearn import datasets
data=datasets.load_iris()
```

```python
data
```

```python
x=pd.DataFrame(data.data,columns=["Sepal length","Sepal width","patel length","patel width"])
```

```python
x
```

```python
data.target
```

```python
y=pd.DataFrame(data.target,columns=["Species"])
```

```python
data=pd.concat([x,y],axis=1)
```

```python
data.to_csv("iris.csv",index=False)
```

```python
import seaborn as sns 
sns.pairplot(data,hue="Species")
```

```python
optimised_data=data[data["Species"]!=0][["Sepal length","Sepal width","Species"]]
```

```python
sns.pairplot(optimised_data,hue="Species")
```

```python
optimised_data.shape 
```

```python
x=optimised_data.iloc[:,0:2]
```

```python
y=optimised_data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
clf3=KNeighborsClassifier()
```

```python
estimators=[('lr',clf1),('dt',clf2),('knn',clf3)]
```

```python
estimators[0]
```

```python
for estimator in estimators:
    X=cross_val_score(estimator[1],x,y,cv=10,scoring='accuracy')
    print(estimator[0],np.round(np.mean(X),2))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
model1=VotingClassifier(estimators=estimators,voting='hard')
```

```python
vca=cross_val_score(model1,x,y,cv=10,scoring='accuracy')
```

```python
np.round(np.mean(vca),2)
```
