```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("heart.csv")
```

```python
df
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()# just keep it here will understand the intition in next class
svc = SVC()
lr = LogisticRegression()
```

```python
#Random forest classifier
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#gradient boost classifier
gb.fit(x_train,y_train)
y_pred = gb.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#Support vector classification 
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#Logistic regression
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#Hyper parameter tuning 
rf2 = RandomForestClassifier(max_samples=0.75,random_state=42)
rf2.fit(x_train,y_train)
y_pred = rf2.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score

print(cross_val_score(RandomForestClassifier(max_samples=0.75),x,y,cv=10,scoring='accuracy'))
print(np.mean(cross_val_score(RandomForestClassifier(max_samples=0.75),x,y,cv=10,scoring='accuracy')))
```

```python
# Number of trees in random forest
n_estimators = [20,60,100,120]

# Number of features to consider at every split
max_features = [0.2,0.6,1.0]

# Maximum number of levels in tree
max_depth = [2,8,None]

# Number of samples
max_samples = [0.5,0.75,1.0]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
              'max_samples':max_samples
             }
print(param_grid)
```

```python
rf3 = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(estimator = rf3, 
                       param_grid = param_grid, 
                       cv = 5, 
                       verbose=2, 
                       n_jobs = -1)

rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
#this is the best paramter 
rf4=RandomForestClassifier(max_depth=2,max_features=0.2,max_samples=0.75,n_estimators=100)
```

```python
rf4.fit(x_train,y_train)
y_pred = rf4.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
rf_grid.best_score_
```

```python
# Number of trees in random forest
n_estimators = [20,60,100,120]

# Number of features to consider at every split
max_features = [0.2,0.6,1.0]

# Maximum number of levels in tree
max_depth = [2,8,None]

# Number of samples
max_samples = [0.5,0.75,1.0]

# Bootstrap samples
bootstrap = [True,False]

# Minimum number of samples required to split a node
min_samples_split = [2, 5]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
              'max_samples':max_samples,
              'bootstrap':bootstrap,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf
             }
print(param_grid)
```

```python
rf5 = RandomForestClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV

rf_grid = RandomizedSearchCV(estimator = rf5, 
                       param_distributions = param_grid, 
                       cv = 5, 
                       verbose=2, 
                       n_jobs = -1)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf6 = RandomForestClassifier(n_estimators=60,min_samples_split=2,min_samples_leaf=1,max_samples=1,max_features=0.2,max_depth=None,bootstrap=True)
```

```python
rf6.fit(x_train,y_train)
y_pred = rf6.predict(x_test)
accuracy_score(y_test,y_pred)
```
